```python
bins = [0, 100000, 300000,float('inf')]
labels = ['Low', 'Mid', 'High']
df['IncomeCategory'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
category_counts = df['IncomeCategory'].value_counts().sort_index()


sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
bar = sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')

for i, val in enumerate(category_counts.values):
    plt.text(i, val + 0.1, str(val), ha='center', va='bottom', fontweight='bold')

plt.title("Income Distribution by Category", fontsize=14)
plt.xlabel("Income Category")
plt.ylabel("Count")
plt.tight_layout
plt.show()
```
